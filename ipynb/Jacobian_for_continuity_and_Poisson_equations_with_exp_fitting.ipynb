### Continuity equation coefficients (interior elements only)

These coefficients are derived for the finite volume method applied to the continuity equation.

```python
h1j, hj, hj1, hm, hp, am, ap, dm, dp = symbols("h_{j-1}, h_{j}, h_{j+1}, h_{-}, h_{+}, a_{-}, a_{+}, d_{-}, d_{+}")
coef_a = 1/hj*(am*hj/(2*hm) + dm/hm)
coef_b = 1/hj*(am*h1j/(2*hm) - ap*hj1/(2*hp) - dm/hm - dp/hp)
coef_c = 1/hj*(-ap*hj/(2*hp) + dp/hp)
coef_a, coef_b, coef_c
```

### Approximating the advection velocity

In these equations we need the advection velocity at the cells faces,

$$
a(x_{j+1/2}) = z^{(i)} \mu^{(i)}(x_{j+1/2})\boldsymbol{E}(x_{j+1/2}) = z^{(i)} \mu^{(i)}(x_{j+1/2}) (-1)\epsilon(x_{j+1/2}) \frac{\partial V}{\partial x}\bigg|_{j+1/2}
$$

where $z^{(i)}$ is the valency of the species $i$ (+1 for holes, -1 for electrons), $\mu^{(i)}(x_{j+1/2})$ and $\boldsymbol{E}(x_{j+1/2})$ are the mobility and the electric field at the forward face. It is convientent to use the short hand,  $\mu^{(i)}(x_{j\pm1/2})=\mu^{(i)}_{\pm}$ to describe quanties evaluated at the forward ($+$) and backward ($-$) face. Next we replace the electric field in terms of the solution variable $V$ which means we now need to make an approximation to the first derivative of $V$ at the face.

$$\frac{\partial V}{\partial x}\bigg|_{j+1/2} = \frac{V_{j+1} - V_{j}}{h_{+}}$$

Therefore we approximate the terms as follows,

$$
a(x_{j+1/2}) = -z^{(i)} \mu^{(i)}_{+} \epsilon_{+} \frac{V_{j+1} - V_{j}}{h_{+}} \\
a(x_{j-1/2}) = -z^{(i)} \mu^{(i)}_{-} \epsilon_{-} \frac{V_{j} - V_{j-1}}{h_{-}}
$$

$\epsilon_{\pm}$ and $\mu_{\pm}$ are a fundemental materials constant (they don't depend on other parameters for the system) therefore how we actually calculate these quanties at the cell faces is completely arbitrary because they do not alter the values of the Jacobian. During implementation we replace these terms with apprporaite linear interpolation functions which are based on the cell values and the cell dimensions.

###Electron specific coefficients

Substitute electron values into the general continuity equation coefficients.

```python
V1j, Vj, Vj1 = symbols('V_{j-1}, V_{j}, V_{j+1}')
epsm, epsp = symbols("epsilon_{-}, epsilon_{+}")


muup = Symbol("\mu^{(u)}_{+}")
muum = Symbol("\mu^{(u)}_{-}")
dup = Symbol("d^{(u)}_{+}")
dum = Symbol("d^{(u)}_{-}")
zu = Symbol("z^{(u)}") # valency

# am and ap functions definition for electrons
#am_subs = zu * muum * (-1) * epsm * (Vj - V1j) / hm
#ap_subs = zu * muup * (-1) * epsp * (Vj1 - Vj) / hp

# Define arbitrary functions to reduce complexity (can subs later)
# Note that the electron and hole velocity is difference because
# it depends on the mobility. But the electron veocity only
# appears in the electron equation and visa-versa. So if we are
# sensible, subs in appropirate value later.
am_subs = Function("am")
ap_subs = Function("ap")
# a_subs = Function("a")

# -- Exponential fitting --
# These are the explicit forms needed for exponential fitting
#kappam = (exp(am_subs * hj / dvm) + 1)/(exp(am_subs * hj / dvm) - 1) - 2/exp(am_subs * hj / dvm)
#kappap = (exp(ap_subs * hj / dvp) + 1)/(exp(ap_subs * hj / dvp) - 1) - 2/exp(ap_subs * hj / dvp)

# Instead of using explicit form, lets define opaque functions to define
# the exponential fitting, we can subs. later, this reduces complexity
kappa = Function("\kappa")
dm_subs = dum + Rational(1,2) * kappa(V1j, Vj) * am_subs(V1j, Vj)
dp_subs = dup + Rational(1,2) * kappa(Vj, Vj1) * ap_subs(Vj, Vj1)
```

```python
ra = coef_a.subs({am : am_subs(V1j,Vj), dm : dm_subs })
rb = coef_b.subs({am : am_subs(V1j,Vj), ap : ap_subs(Vj,Vj1), dm : dm_subs, dp : dp_subs})
rc = coef_c.subs({ap : ap_subs(Vj,Vj1), dp : dp_subs})
ra, rb, rc
```

###Hole specific coefficients

Substitute hole specific values into the general continuity equation coefficients.

```python
muvp = Symbol("\mu^{(v)}_{+}")
muvm = Symbol("\mu^{(v)}_{-}")
dvp = Symbol("d^{(v)}_{+}")
dvm = Symbol("d^{(v)}_{-}")
zv = Symbol("z^{(v)}") # valency

# am and ap functions definition for holes
#am_subs = zv * muvm * (-1) * epsm * (Vj - V1j) / hm
#ap_subs = zv * muvp * (-1) * epsp * (Vj1 - Vj) / hp


# Exponential fitting (kappa=0 central; kappa=1 upwind)

#kappam = (exp(am_subs * hj / dvm) + 1)/(exp(am_subs * hj / dvm) - 1) - 2/exp(am_subs * hj / dvm)
#kappap = (exp(ap_subs * hj / dvp) + 1)/(exp(ap_subs * hj / dvp) - 1) - 2/exp(ap_subs * hj / dvp)
dm_subs = dvm + Rational(1,2) * kappa(V1j, Vj) * am_subs(V1j, Vj)
dp_subs = dvp + Rational(1,2) * kappa(Vj, Vj1) * ap_subs(Vj, Vj1)
#am_subs, ap_subs, dm_subs, dp_subs
```

```python
sa = coef_a.subs({am : am_subs(V1j,Vj), dm : dm_subs })
sb = coef_b.subs({am : am_subs(V1j,Vj), ap : ap_subs(Vj,Vj1), dm : dm_subs, dp : dp_subs})
sc = coef_c.subs({ap : ap_subs(Vj,Vj1), dp : dp_subs})
sa, sb, sc
```

###Poisson equation

```python
u1j, uj, uj1 = symbols("u_{j-1}, u_j, u_{j+1}")
v1j, vj, vj1 = symbols("v_{j-1}, v_j, v_{j+1}")
la = 1/hj * epsm/hm
lb = -1/hj * (epsm/hm + epsp/hp)
lc = 1/hj * epsp / hp
#eps1j, epsj, epsj1 = symbols("epsilon_{j-1}, epsilon_{j}, epsilon_{j+1}")
#epsp_subs = (hj1 / (2*hp) * epsj  + hj /(2*hp)*epsj1)
#epsm_subs = (hj  / (2*hm) * eps1j + h1j/(2*hm)*epsj)
#la = la.subs({epsm : epsm_subs})
#lb = lb.subs({epsm : epsm_subs, epsp : epsp_subs})
#lc = lc.subs({epsp : epsp_subs})
Nj = Symbol("N_j")
rho = zu*uj + zv*vj + Nj
poisson = la * V1j + lb * Vj + lc * Vj1 + rho
poisson
```

### Continuity equations

```python
electron = ra * u1j + rb * uj + rc * uj1
holes = sa * v1j + sb * vj + sc * vj1
electron
```

### Complete system of equations

```python
A = Matrix([[ra, 0, 0, rb, 0, 0, rc, 0, 0],
            [0, sa, 0, 0, sb, 0, 0, sc, 0],
            [0,  0,la, 0,  0, lb, 0,  0,lc]])

R = Function("R") # Reaction
d = Matrix([R(uj,vj), R(uj,vj), rho])
X = Matrix([u1j, v1j, V1j, uj, vj, Vj, uj1, vj1, Vj1])

F = A*X + d
F
```
