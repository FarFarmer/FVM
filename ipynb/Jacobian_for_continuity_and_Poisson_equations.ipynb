### Continuity equation coefficients (interior elements only)

These coefficients are derived for the finite volume method applied to the continuity equation.

```python
h1j, hj, hj1, hm, hp, am, ap, dm, dp = symbols("h_{j-1}, h_{j}, h_{j+1}, h_{-}, h_{+}, a_{-}, a_{+}, d_{-}, d_{+}")
coef_a = 1/hj*(am*hj/(2*hm) + dm/hm)
coef_b = 1/hj*(am*h1j/(2*hm) - ap*hj1/(2*hp) - dm/hm - dp/hp)
coef_c = 1/hj*(-ap*hj/(2*hp) + dp/hp)
coef_a, coef_b, coef_c
```

### Approximating the advection velocity

In these equations we need the advection velocity at the cells faces,

$$
a(x_{j+1/2}) = z^{(i)} \mu^{(i)}\boldsymbol{E}_{j+1/2} = z^{(i)} \mu^{(i)} (-1)\epsilon(x_{j+1/2}) \frac{\partial V}{\partial x}\bigg|_{j+1/2}
$$

where $z^{(i)}$ is the valency of the species $i$ (+1 for holes, -1 for electrons), $\mu^{(i)}$ is the mobility, and $\boldsymbol{E}_{j+1/2}$ is the electric field at the face. We replace the electric field in terms of the solution variable $V$ which means we now need to make an approximation to the first derivative of $V$ at the face.

$$\frac{\partial V}{\partial x}\bigg|_{j+1/2} = \frac{V_{j+1} - V_{j}}{h_{+}}$$

Therefore we approximate the terms as follows,

$$
a(x_{j+1/2}) = z_i \mu_i (-1) \epsilon(x_{j+1/2}) \frac{V_{j+1} - V_{j}}{h_{+}} \\
a(x_{j-1/2}) = z_i \mu_i (-1) \epsilon(x_{j-1/2}) \frac{V_{j} - V_{j-1}}{h_{-}}
$$

$\epsilon(x_{j-1/2})$ and $\epsilon(x_{j+1/2})$ is a fundemental materials constant and as such doesn't depend on any other variable in the system, therefore how we actually calculate $\epsilon$ at the faces is a completley free choice. During implementation we replace these terms with apprporaite linear interpolation functions which are based on the cell values and the cell dimensions.

###Electron specific coefficients

Substitute electron values into the general continuity equation coefficients.

```python
V1j, Vj, Vj1 = symbols('V_{j-1}, V_{j}, V_{j+1}')
epsm, epsp = symbols("epsilon_{-}, epsilon_{+}")
muup = Symbol("\mu^{(u)}_{+}")
muum = Symbol("\mu^{(u)}_{-}")
dup = Symbol("d^{(u)}_{+}")
dum = Symbol("d^{(u)}_{-}")
zu = Symbol("z^{(u)}") # valency of electrons
ap_subs = zu * muup * (-1) * epsp * (Vj1 - Vj) / hp
am_subs = zu * muum * (-1) * epsm * (Vj - V1j) / hp
dm_subs = dum
dp_subs = dup
am_subs, ap_subs, dm_subs, dp_subs
```

```python
ra = coef_a.subs({am : am_subs, dm : dm_subs})
rb = coef_b.subs({am : am_subs, ap : ap_subs, dm : dm_subs, dp : dp_subs})
rc = coef_c.subs({ap : ap_subs, dp : dp_subs})
ra, rb, rc
```

###Hole specific coefficients

Substitute hole specific values into the general continuity equation coefficients.

```python
muvp = Symbol("\mu^{(v)}_{+}")
muvm = Symbol("\mu^{(v)}_{-}")
dvp = Symbol("d^{(v)}_{+}")
dvm = Symbol("d^{(v)}_{-}")
zv = Symbol("z^{(v)}") # valency of electrons
ap_subs = zv * muvp * (-1) * epsp * (Vj1 - Vj) / hp
am_subs = zv * muvm * (-1) * epsm * (Vj - V1j) / hp
dm_subs = dvm
dp_subs = dvp
am_subs, ap_subs, dm_subs, dp_subs
```

```python
sa = coef_a.subs({am : am_subs, dm : dm_subs})
sb = coef_b.subs({am : am_subs, ap : ap_subs, dm : dm_subs, dp : dp_subs})
sc = coef_c.subs({ap : ap_subs, dp : dp_subs})
sa, sb, sc
```

###Poisson equation

```python
u1j, uj, uj1 = symbols("u_{1-j}, u_j, u_{j+1}")
v1j, vj, vj1 = symbols("v_{1-j}, v_j, v_{j+1}")
la = 1/hj * epsm/hm
lb = -1/hj * (epsm/hm + epsp/hp)
lc = 1/hj * epsp / hp
#eps1j, epsj, epsj1 = symbols("epsilon_{j-1}, epsilon_{j}, epsilon_{j+1}")
#epsp_subs = (hj1 / (2*hp) * epsj  + hj /(2*hp)*epsj1)
#epsm_subs = (hj  / (2*hm) * eps1j + h1j/(2*hm)*epsj)
#la = la.subs({epsm : epsm_subs})
#lb = lb.subs({epsm : epsm_subs, epsp : epsp_subs})
#lc = lc.subs({epsp : epsp_subs})
Nj = Symbol("N_j")
rho = zu*uj + zv*vj + Nj
poisson = la * V1j + lb * Vj + lc * Vj1 + rho
poisson
```

### Continuity equations

```python
electron = ra * u1j + rb * uj + rc * uj1
holes = sa * v1j + sb * vj + sc * vj1
electron
```

```python
holes
```

### Jacobians

```python
X = Matrix([electron, holes, poisson])
Y = Matrix([u1j, v1j, V1j])
X.jacobian(Y)
```

```python
Y = Matrix([uj, vj, Vj])
X.jacobian(Y)
```

```python
Y = Matrix([uj1, vj1, Vj1])
X.jacobian(Y)
```

### Complete system of equations

```python
A = Matrix([[ra, 0, 0, rb, 0, 0, rc, 0, 0],
            [0, sa, 0, 0, sb, 0, 0, sc, 0],
            [0,  0,la, 0,  0, lb, 0,  0,lc]])

d = Matrix([0,0,rho])

X = Matrix([u1j, v1j, V1j, uj, vj, Vj, uj1, vj1, Vj1])

F = A*X + d
F
```

### Jacobian of complete system (excluding boundary conditions)

```python
J = F.jacobian(X)
J
```
